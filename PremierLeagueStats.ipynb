```python
import pandas as pd


df = pd.read_csv('dataset - 2020-09-24.csv')

import plotly.io as pio
pio.renderers.default = 'colab'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


df.head()
```

```python

df.info()

print(df.columns)

df.describe()
```

```python
fig = px.pie(df,
             values='Appearances',
             names='Nationality',
             title='Countries Represented in the EPL by Number of Appearances')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='<b> Number of Appearances per Country </b>',
                  title_x=0.5,
                  titlefont=dict(size=18, family="Arial"),
                  width=600, height=600, showlegend=False)
fig.show()
```

```python
top_countries = df.groupby('Nationality')['Appearances'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries by Appearances')
plt.xlabel('Number of Appearances')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Appearances', y='Goals', data=df)
plt.title('Appearances vs Goals')
plt.xlabel('Appearances')
plt.ylabel('Goals')
plt.show()
```

```python
fig = px.histogram(df,
                   x='Goals',
                   nbins=20,
                   title='Distribution of Goals Scored')
fig.update_layout(xaxis_title='Goals', yaxis_title='Frequency')
fig.show()
```
